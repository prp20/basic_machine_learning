# Exploratory Data Analysis - Bivariate Analysis

```python
#load multiple datasets
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

### a. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
#Multivariate Analysis
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

### b. Barplot (Numerical - Categorical)

```python
sns.barplot(data=titanic, x='pclass', y='age')
```

```python
sns.barplot(data=titanic, x='pclass', y='fare', hue='sex')
```

```python
sns.barplot(data=titanic, x='pclass', y='age', hue='sex')
```

### c. BoxPlot (Numerical - Categorical)

```python
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
```

### d. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'], hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'], hist=False)
```

### e. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['sex']))
```

 ### f. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```

### g. Pairplot (Numerical - Numerical)

```python
iris.head()
```

```python
sns.pairplot(data=iris, hue='species')
```

### h. Lineplot (Numerical - Numerical) (to be used only when x axis is be time based)

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(data=new, x='year', y='passengers')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
